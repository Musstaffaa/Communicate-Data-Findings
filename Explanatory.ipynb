## How much damage is there in Egypt compared to the countries that have reported the most against violence?

```python
fig, ax =plt.subplots(figsize=(10,7))
ax.hist(df2['Value'], alpha=0.6, label= '400 cases country')
ax.hist(Egypt['Value'], alpha=0.6, label= 'Egypt')
ax.set_xlabel('Damage distribution')
ax.set_ylabel('count')
ax.legend(loc='upper right')
ax.set_title('A comparison of harm in the study of Egypt and the countries with the most reporting of violence')

```

- And very sadly, we see that Egypt has a great deal of harm to women, as when the amount of harm in Egypt was compared only with the rest of the countries, we see that Egypt has a very, very great harm to women, or in general, violence against women.
